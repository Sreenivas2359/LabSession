* From the Chg% (available in column I) find the probability of making a loss over the stock. 

```python
q['classify'] = ''
q
```

```python
q['classify'] = q['Chg%'].apply(lambda x: 'PROFIT' if x > 0  else 'LOSS');
q
```

```python
# Total amount of profit over stocks
q1=q.loc[q['classify']=="PROFIT"]
q1
```

```python
Total_profit = len(q1.axes[0])
Total_profit
```

```python
# Total amount of loss over stocks

q2=q.loc[q['classify']=="LOSS"]
q2
```

```python
Total_loss = len(q2.axes[0])
Total_loss
```

```python
# probability of making a loss over the stock. 

prob_loss = np.divide(Total_loss,(Total_profit + Total_loss))
prob_loss
```
